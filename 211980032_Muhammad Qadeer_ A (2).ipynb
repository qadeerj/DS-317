# nunique()

The nunique() method returns the number of unique values for each column.

By specifying the column axis (axis='columns'), the nunique() method searches column-wise and returns the number of unique values for each row.



```python
df.nunique()
```

# isnull()

The isnull() method returns a DataFrame object where all the values are replaced with a Boolean value True for NULL values, and otherwise False.

```python
df.isnull()
```

# fillna()

The fillna() method in Python is used to replace missing or null values in a DataFrame or Series with a specified value. The method returns a new DataFrame object unless the inplace parameter is set to True, in which case the fillna() method replaces the missing values in the original DataFrame
In this example, we create a DataFrame with null values in columns 'A' and 'B'. We then use fillna() to replace the null values with 0. The inplace=True parameter modifies the original DataFrame instead of returning a new one

```python
df.fillna(0, inplace=True)
df
```

# sort_values()

The sort_values() method in Python Pandas is used to sort a DataFrame by one or more columns in ascending or descending order. The method takes the column name(s) as an argument and returns a new sorted DataFrame object unless the inplace parameter is set to True, in which case the method modifies the original DataFrame

```python
df.sort_values('Delivery_Time(Days)') # for asending
```

We can also sort a DataFrame by multiple columns by passing a list of column names to sort_values()

```python
df.sort_values(['Delivery_Time(Days)','Shipping_Cost(USD)'],ascending=[True, False])
```

In this example, we create a DataFrame with two column. We then use sort_values() to sort the DataFrame by column 'Delivery_Time(Days)' in ascending order, then by column 'Shipping_Cost(USD)' in descending order. The ascending parameter is set to [True, False] to specify the sort order for each column.

# value_counts()

The value_counts() method in Python Pandas is used to count the unique occurrences of values in a Series or DataFrame column. The method returns a new Series object containing the counts of unique values in descending order, so that the first element is the most frequently-occurring element. By default, the method excludes null values

```python
res = df.value_counts('Shipping_Cost(USD)')
res
```

# nlargest()

The nlargest() method in Python Pandas is used to get the n largest values from a DataFrame or a Series. The method returns a new DataFrame or Series object containing the n largest values, sorted in descending order. By default, the method returns the first occurrence of a duplicate value.We can also use nlargest() to get the n largest values from a DataFrame column. 

```python
res = df.nlargest(20,'Quantity')
res
```

# nsmallest()

The nsmallest() method in Python Pandas is used to get the n smallest values from a DataFrame or a Series. The method returns a new DataFrame or Series object containing the n smallest values, sorted in ascending order. By default, the method returns the first occurrence of a duplicate value
.We can also use nsmallest() to get the n Samllest values from a DataFrame column. 

```python
df.nsmallest(500,'Quantity')
```

# df.copy()

The copy() method in Python Pandas is used to create a copy of a DataFrame or a Series. By default, it creates a deep copy, meaning that the new DataFrame or Series is independent of the original one, and modifications to the copy will not affect the original DataFrame or Series

```python
df_copy = df.copy()

# modify the copy
df_copy.loc[0, 'Status'] = 'Failed'

# print the original DataFrame and the copy
print('Original DataFrame:')
print(df)
print('Copied DataFrame:')
print(df_copy)
```

# df.rename()

# df.where()

In Python Pandas, the where() method is used to replace values in a DataFrame based on a condition. The method returns a new DataFrame with the same shape as the original one, where the values that meet the condition are replaced with the specified value, and the values that do not meet the condition are left unchanged


```python
df['Quantity'] = df['Quantity'].where(df['Quantity'] == 92, other=0)
df
```
